```python
from urllib import request
import pandas as  pd
import numpy as np
```

**1. Download do Dataset do Portal de Dados Abertos**





```python
file_url = 'https://ons-dl-prod-opendata.s3.amazonaws.com/dataset/carga_energia_di/CARGA_ENERGIA_2022.csv'
source_file = 'CARGA_ENERGIA_2022.csv'
request.urlretrieve(file_url , source_file )
```

**2. Importar arquivo**

```python
dataframe = pd.read_csv(source_file, sep=';',encoding = "utf-8")

if (dataframe.empty) :
    print('DATAFRAME VAZIO')
else:
  print('shape:', dataframe.shape)
  print('columns:', dataframe.columns)
```

**5. Análise sobre os resultados - Foco em Geração**

*A função value_counts() do pandas é valiosíssima nas análises de dados, o que ela faz é verificar a contagem de dados iguais em um dataframe ou em uma coluna, e retorna uma Series com esse resultado.*

```python
dataframe.id_subsistema.value_counts(ascending=False)
```

*A função duplicated() nos retorna uma série booleana denotando linhas duplicadas. Os resultados são Verdadeiro(True) e Falso(False), sendo verdadeiros os valores duplicados.*

```python
resultado = dataframe.duplicated(subset=['id_subsistema'], keep=False)
dataframe[resultado]
```
